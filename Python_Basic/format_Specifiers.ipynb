```python
# format specifiers = {value:flags} fromat a value based on 
#                      what flags are inserted

# .(number)f = round to that many decimal places (fixed point)
# :(number) = allocate that many spaces
# :03 = allocate and Zero pad that many space
# :< = left justify
# :> = right justify
# :^ = center align
# :+ = use a plus sign to indicate positive value
# := = place sign to leftmost position
# : = insert a space before positive numbers
# :, = comma separator

price1 = 3.14159
price11 = 352525.14159
price12 = 3235252.14159
price2 = -987.65
price3 = 12.34

#for num to be print after decimal
print (f"Price 1 is ${price1:.2f}")
print (f"Price 2 is ${price2:.2f}")
print (f"Price 3 is ${price3:.2f}")
print("-----------------------------------")
# for spacing
print (f"Price 1 is ${price1:10}")
print (f"Price 2 is ${price2:10}")
print (f"Price 3 is ${price3:10}")
print("-----------------------------------")

# Each no. is now 0 padded
print (f"Price 1 is ${price1:010}")
print (f"Price 2 is ${price2:010}")
print (f"Price 3 is ${price3:010}")
print("-----------------------------------")

# all the no. are Right justify
print (f"Price 1 is ${price1:>10}")
print (f"Price 2 is ${price2:>10}")
print (f"Price 3 is ${price3:>10}")
print("-----------------------------------")

# all the no. are left justify
print (f"Price 1 is ${price1:<10}")
print (f"Price 2 is ${price2:<10}")
print (f"Price 3 is ${price3:<10}")
print("-----------------------------------")

# all the no. are Aline Center
print (f"Price 1 is ${price1:^10}")
print (f"Price 2 is ${price2:^10}")
print (f"Price 3 is ${price3:^10}")
print("-----------------------------------")

# all the no. are Aline Center
print (f"Price 1 is ${price11:,}")
print (f"Price 2 is ${price12:,}")
print (f"Price 3 is ${price3:,}")


```
